```python
# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
```

```python
# 2. Load dataset from a CSV file
# Replace 'your_file.csv' with the actual filename
df = pd.read_csv("/content/iris (2).csv")
```

```python
# Show all column names in the dataset
print(df.columns)
```

```python
# 3. Split features (X) and target (y)
X = df.drop("species", axis=1)
y = df["species"]
```

```python
# 4. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 5. Build the Decision Tree (with pruning: max_depth = 3)
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X_train, y_train)
```

```python
# 6. Visualize the tree
plt.figure(figsize=(12,8))
plot_tree(clf, feature_names=X.columns, class_names=[str(c) for c in y.unique()], filled=True)
plt.show()
```

```python
# 7. Predictions
y_pred = clf.predict(X_test)
```

```python
# 8. Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1-score:", f1_score(y_test, y_pred, average="weighted"))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
